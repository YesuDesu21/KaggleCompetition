Row format for csv file submission:
PassengerId - sorted any order
Survived - binary prediction (1- survived; 0 - deceased)

Problem:
1. Predict if passenger survived or not
2. Each test set predict 0 or 1 value for variable

Use gender as basis and survived as output (y)

Output file:
Depending on the gender, display person's id in output.

```python
import pandas as pd

train_data = pd.read_csv("../titanic/train.csv")
test_data = pd.read_csv("../titanic/test.csv")

test_id = test_data["PassengerId"]
```

```python
#Preprocess
def clean(data):
    
    data = data.drop(["Ticket","PassengerId","Name","Cabin"], axis=1)

    column = ["SibSp","Parch","Fare","Age"]

    for col in column:
        data[col].fillna(data[col].median(), inplace=True)
    
    #Fill in unknown with U
    data.Embarked.fillna("U", inplace=True)

    return data

train_data = clean(train_data)
test_data = clean(test_data)
```

```python
from sklearn import preprocessing

#Category to num?
le = preprocessing.LabelEncoder()
columns = ["Sex", "Embarked"]

for cols in columns:
    train_data[cols] = le.fit_transform(train_data[cols])
    test_data[cols] = le.transform(test_data[cols])
    print(le.classes_)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

#You can do linear, KNN, decision tree, etc
y = train_data["Survived"]
x = train_data.drop("Survived", axis=1)

#ran.state = 42 for reproductability 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
classifier = LogisticRegression(random_state = 0, max_iter=1000).fit(x_train, y_train)
```

```python
prediction = classifier.predict(x_test)

from sklearn.metrics import accuracy_score
accuracy_score(y_test, prediction)
```

```python
submission = classifier.predict(test_data)
```

```python
df = pd.DataFrame({"PassengerId":test_id.values,
                    "Survived": submission
                })
```

```python
#Index to false     
df.to_csv("submission.csv", index=False)
```
